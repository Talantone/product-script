```python
# import necessary libraries
import os
import glob
import dask.dataframe as dd
import datetime
from dateutil.relativedelta import relativedelta
  
path = os.getcwd()
new_path = os.path.join(path, "inp")

df = dd.read_csv(new_path + "/*.csv", sep=";", header=None)
df.columns = ['Variable symbol', 'Creation date', 'Due date', 'Amount', 'VAT', 'Payment type', 'Is payed']
df['Creation date'] = dd.to_datetime(df['Creation date'])
df = df.set_index(['Creation date'])
```

```python
timestamp = datetime.datetime.now()
datetime_before_month = timestamp - relativedelta(months=1)
df1 = df.loc[datetime_before_month:timestamp]
df1.compute()
```
